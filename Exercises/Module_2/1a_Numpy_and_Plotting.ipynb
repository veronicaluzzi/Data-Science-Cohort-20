## Problem 1

Generate 10,000 samples from a normal distribution with mean 0 and standard deviation 1.

````python
import numpy as np
import pandas as pd

# Set a seed for reproducibility
np.random.seed(42)

# Generate 10,000 samples from a normal distribution
# loc: mean, scale: standard deviation, size: number of samples
samples = np.random.normal(loc=0, scale=1, size=10000)

# Create a DataFrame to store the samples
df = pd.DataFrame(samples, columns=['Value'])

# Save the samples to a CSV file
df.to_csv('normal_distribution_samples.csv', index=False)

# Display the first few samples
print(df.head())
````

## Problem 3

Write a for loop that loops through the mean values in `mean_list`.

*   In each loop, generate 10,000 samples from a normal distribution with mean `i` (the value in `mean_list`) and standard deviation 1. For example, loop 1 will generate samples from a normal distribution with mean 0, loop 2 will generate samples from a normal distribution with mean 2, and so on.

*   In each loop, also plot a histogram of the samples you generate. Plot the histograms on the same figure (i.e., do not use subplots). Set `alpha` = 0.5 so that the histograms are somewhat transparent.
*   Add a legend and axis labels.


````python
import numpy as np
import matplotlib.pyplot as plt

# Define the list of means
mean_list = [0, 2, 4, 6]
std_dev = 1
num_samples = 10000

# Set a seed for consistent results
np.random.seed(42)

# Loop through each mean in the list
for i in mean_list:
    # Generate 10,000 samples for the current mean 'i'
    samples = np.random.normal(loc=i, scale=std_dev, size=num_samples)

    # Plot the histogram on the same figure
    # bins=50 creates enough detail, alpha=0.5 makes bars transparent
    plt.hist(samples, bins=50, alpha=0.5, label=f'Mean = {i}')

# Add styling and labels
plt.title('Comparison of Normal Distributions with Different Means ($\mu$)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()  # Show the legend to identify each mean

# Save or display the plot
plt.savefig('histograms_comparison.png')
````

## Problem 5

Create a plot of `free_time` vs `age` with these additional features:
- Use both markers and a line
- Change your marker shape to be a triangle
- Change the colors of the line and the markers to colors of your choice
- Change the marker size and the linewidth so they are easier to see
- Add axis labels


````python
import numpy as np
import matplotlib.pyplot as plt

# Create the data
age = np.linspace(1, 100, 20)
free_time = (age - 50)**2

# Create the plot
plt.plot(age, free_time,
         marker='*',               # Triangle markers
         linestyle='-',            # Solid line
         color='blue',       # Line color
         markerfacecolor='gold',    # Marker fill color
         markersize=13,            # Marker size
         linewidth=2.5)            # Line width

# Add axis labels and title
plt.xlabel('Age')
plt.ylabel('Free Time')
plt.title('Free Time vs Age')

# Save the plot
plt.savefig('free_time_plot.png')
````

## **BONUS** (only if you have time)

Write a function that takes in a string as its input. This function should return the number of letters and words in the string.

For example:

string_counter("Hello, my name is Bob.") -> (16, 5)

string_counter("What's up?") -> (7, 2)

string_counter(" ") -> (0,0)

```python
string_counter('Hello, my name is Bob.')

string_counter("What's up?")

string_counter(' ')
```



````python
def string_counter(input_string):
    # Count only characters that are letters (a-z, A-Z)
    letters = sum(c.isalpha() for c in input_string)

    # Split the string by whitespace and filter out empty strings
    words_list = input_string.split()
    words = len(words_list)

    return (letters, words)

# Test cases
print(string_counter("My dog's name is Sage")) # Expected: (16, 5)
print(string_counter("The sun is up?"))              # Expected: (7, 2)
print(string_counter(" "))                      # Expected: (0, 0)
````
