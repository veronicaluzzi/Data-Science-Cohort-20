Build each of the following games in Python:

1. **Rock, Paper, Scissors** - In this game, the user will be playing Rock, Paper, Scissors against the computer. ## Rock, Paper, Scissors

```python
import random

# 1. Define the options: create a list
options = ["rock", "paper", "scissors"]

# 2. Get the computer's choice: created a variable
computer_choice = random.choice(options)

# 3. Get the user's choice: create another variable for the user.  Use '.lower' to always have the user's choice in lower font.
user_choice = input("Enter rock, paper, or scissors: ").lower()

# 4. Check for a tie: use if/elif to check if the user and computer choice are identical.
if user_choice == computer_choice:
    print(f"Both players selected {user_choice}. It's a tie!")

# 5. Determine the winner using conditional logic
elif user_choice == "rock":
    if computer_choice == "scissors":
        print("You win! Rock smashes scissors.")
    else:
        print("The computer won. Paper covers rock.")

elif user_choice == "paper":
    if computer_choice == "rock":
        print("You win! Paper covers rock.")
    else:
        print("The computer won. Scissors cuts paper.")

elif user_choice == "scissors":
    if computer_choice == "paper":
        print("You win! Scissors cuts paper.")
    else:
        print("The computer won. Rock smashes scissors.")

# Handle invalid input
else:
    print("Invalid input! Please check your spelling.")
```

### **How it Works**
## import random:

This is a built-in Python module. We use random.choice(options) to let the computer pick a random string from our list.

## .lower():

Users are unpredictable. They might type "Rock", "ROCK", or "rock". Adding .lower() converts their input to lowercase so it always matches our list.

## Nested if statements:

We first check the user's choice, and then check the computer's choice inside that block to decide the winner.

## Why Logic Order Matters


In programming, we usually handle the Tie first. If the choices are the same, we don't need to check any other rules, which makes the code more efficient.

## Pro Tip:

If you want to make this even better, you could wrap the whole thing in a while True: loop so you can play multiple rounds without restarting the script!

## MadLibs

```python
# 1. Collect inputs from the user
print("--- Welcome to MadLibs! ---")
adj1 = input("Enter an adjective: ")
noun1 = input("Enter a noun: ")
verb1 = input("Enter a verb (past tense): ")
place = input("Enter a place: ")
adj2 = input("Enter another adjective: ")
noun2 = input("Enter one more noun: ")

# 2. Format the inputs
# Using .strip() removes accidental spaces, .lower() keeps it consistent
noun1 = noun1.strip().lower()
adj1 = adj1.strip().lower()

# 3. Create the story using an f-string
story = f"""
Today was a very {adj1} day!
I decided to take my darling {noun1} and {verb1} all the way to the {place.title()}.
When I arrived, I saw a {adj2} {noun2} waiting for us.
It was scary but we figured out a path around the magnificent beast!
"""

# 4. Print the final result
print("\n--- Your MadLib Story ---")
print(story)
```

### Why this works well:


## Triple Quotes (f"""):

Using three sets of quotes allows you to write a multi-line story exactly how you want it to look, without needing a bunch of separate print() statements.

## .strip():

Beginners often accidentally hit the spacebar before or after typing a word. .strip() cleans that up so your story doesn't look like: "I saw a  dog  waiting for me."

## .title():

For the place variable, I used .title() so that even if the user types "albuquerque," it prints as "Albuquerque."

## The \n:

This adds a "newline" (a blank line) before the story prints, making it easier to read in the console.

## Number Guesser

Number Guesser - This is a simple guessing game where the computer will generate a random number between 1 to 10, and the user has to guess it in 5 attempts. Based on the user’s guess, the computer will give a hint if the number is high or low. When the user correctly guesses, print the answer along with the number of attempts.

```python
import random

# 1. Setup the game
secret_number = random.randint(1, 10)
attempts_allowed = 5
attempts_taken = 0

print("--- I'm thinking of a number between 1 and 10 ---")
print(f"You have {attempts_allowed} attempts to guess it.")

# 2. Start the guessing loop
while attempts_taken < attempts_allowed:
    # Get the user's guess
    guess = int(input(f"\nAttempt {attempts_taken + 1}: Enter your guess: "))
    attempts_taken += 1  # Increment the counter

    # 3. Check the guess
    if guess == secret_number:
        print(f"Correct! The answer was {secret_number}.")
        print(f"It took you {attempts_taken} attempts to win!")
        break  # This exits the loop early because they won
    elif guess < secret_number:
        print("Too low! Try a higher number.")
    else:
        print("Too high! Try a lower number.")

# 4. If the loop finishes without a correct guess
else:
    print(f"\nGame Over! You've used all {attempts_allowed} attempts.")
    print(f"The number I was thinking of was {secret_number}.")
```

### Key Concepts Used


## random.randint(1, 10):

This generates a whole number between 1 and 10, inclusive.

## int(input(...)):

The input() function always returns text (a string). Since we are comparing numbers, we have to "wrap" it in int() to convert it to an integer.

## while attempts_taken < attempts_allowed:

This loop keeps running as long as the player has guesses left.

## The break Keyword:

This is very useful. It tells Python to stop the loop immediately once the user gets the right answer, even if they have 3 tries left.

## while / else:

This is a cool Python feature. The else block after a loop only runs if the loop finished naturally (i.e., the user ran out of guesses) and was not stopped by a break.

## Limitations:

If the user types a string, this approach will not work.  I will need to use something called: try/except block.
