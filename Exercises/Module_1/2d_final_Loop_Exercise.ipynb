## Problem 2



Write a Python program that takes a word and reverses it.

Input:

```
input_word = 'stressed'
```

Output:
```
desserts
```

Solution:

````python
#create a word
word='Veronica'
word
````

````python
#reverse the word
word = "veronica"
#python by default adds a (\n) character at the end of each letter.
#to avoid that character and have the reverse word be spell out on a line, add (end="") to the code.
for rev in word[::-1]:
    print(rev, end="")
````

- f-strings (f"..."): This is a modern way to format strings in Python. It allows you to put variables directly inside the text using curly braces {}.

# Problem 5



Write a Python program which iterates through the integers from 0 to 50. For multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".  

For this problem zero is considered a multiple of three and five 3\*0=5\*0=0

Sample Output :  
```
Fizz Buzz
1
2
Fizz
4
Buzz
Fizz
7
8
Fizz
Buzz
<etc.>
```


````python
#creating a first attempt
for i in range(0, 51,3):
    # Check for both first
    if i % 3 == 0 and i % 5 == 0:
        print(f"{i}: FizzBuzz")
    elif i % 3 == 0:
        print(f"{i}: Fizz")
    elif i % 5 == 0:
        print(f"{i}: Buzz")
````

````python
#because the numbers in the range above were all multiple of 3, I deleted the step information
for i in range(0, 51):
    # Check for both first
    if i % 3 == 0 and i % 5 == 0:
        print(f"{i}: FizzBuzz")
    elif i % 3 == 0:
        print(f"{i}: Fizz")
    elif i % 5 == 0:
        print(f"{i}: Buzz")
````

````python
#cleaning up and having it printing all the numbers not only the ones that are multiple of 3 and 5.
for i in range(0, 51):
    # Check if number is divisible by both 3 and 5
    if i % 3 == 0 and i % 5 == 0:
        print("Fizz Buzz")
    # Check if number is divisible by 3
    elif i % 3 == 0:
        print("Fizz")
    # Check if number is divisible by 5
    elif i % 5 == 0:
        print("Buzz")
    # If none of the above, just print the number
    else:
        print(i)
````
