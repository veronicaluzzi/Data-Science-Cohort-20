1. Given	a list,	remove the	element	at	index	number 4 and add	it at index position 2 (over-writing what is currently there) and also add it at the end of the list.

```python

```

````
#attempt to remove [4]. first create and display list.
list1 = [34, 56, 49, 4900, 249, 494]
list1
````

````
# 1. Remove element at index 4 and store it in a variable and
# created a new variable called pop1 where the popped element 249 gets stored.
pop1 = list1.pop(4)
pop1
````

````
list1
````

````
#remove the element at index number 4 and add it at index position 2 (over-writing what is currently there) and also add it at the end of the list

# 2. Insert that value into index 2
list1.insert(2, pop1)
list1

````

````
# 3. Add that same value to the end of the list
list1.append(pop1)
list1
````
