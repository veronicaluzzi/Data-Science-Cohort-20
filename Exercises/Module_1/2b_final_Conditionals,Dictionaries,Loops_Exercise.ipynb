5. Write a python program that will loop through the following list:
`pets = ['Muffin', 'Kiki', 'Moe', 'Daisy']`. If the length of the pet's name is less than 5 characters, print out the name.  

````python
#first try
#If the length of the pet's name is less than 5 characters, print out the name.
pets = ['Muffin', 'Kiki', 'Moe', 'Daisy']
pets
````

````python
#use the lecture on loops, and the concept of 'len'
pets = ['Muffin', 'Kiki', 'Moe', 'Daisy']
for pet in pets:
    # len(pet) counts how many characters are in the name
    if len(pet) < 5:
        print(pet)
````
