2. Create a variable called `nested` and set it equal to the following nested list: `[[1,2,3], ['a','b','c']]`. Use indexing to access the `b` from your nested list. Use indexing to access `[2,3]` from your nested list.

```python
# first I created the nested list that contains the 2 inner lists indicated in the instructions.  I found this method in Google by asking 'how to access an element in a list within a list in python'.
#The response was using Basic Method: Successive Indexing and that is here: https://share.google/aimode/WcdYbfzVAKyTflXAy
nested_list = [[1,2,3], ['a','b','c']]
nested_list
```

```python
# then, I assign the 'nested' list to be = to the original 'nested_list'
nested = nested_list
nested
```

```python
# now, I will specifically name the list inside the list that contains the 'b'
second_inner_list = nested[1]
print (second_inner_list)
```

```python
# in this step, I am trying to access the 'b' element within the 'second_inner-list'
my_element = nested_list[1][1]
print(my_element)
```

```python
#finally, i am creating a name for the list inside the list that contains [2,3]
first_inner_list = nested[0]
print (first_inner_list)
```

```python
my_second_element = first_inner_list [1::1]
print(my_second_element)
```
