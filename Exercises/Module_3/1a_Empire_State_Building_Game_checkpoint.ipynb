#Climbing The Steps of The Empire State Building Steps: The Game
The objective is to create a Dice game logic, which will decide the odds of reaching Step number 60 of the Empire State Building in 100 rolls of the die. We would then simulate this sequence multiple times to calculate the probability of reaching step 60 in 100 tries using the Dice game logic.


#Rules
For each random walk sequence we throw a Die.
1. Step 0 at ground level, is the starting point.
2. If the roll of the Die returns 1 or 2, climb down one step.
3. If the roll of the Die returns 3, 4 or 5, go up one step.
4. If the roll of the Die returns 6, throw the Die again and advance the number of steps equal to that throw.
5. There is a little chance of being clumsy and falling down the Stairs in which case we start again from 0.
6. The value of Steps cannot be negative, the lowest value would always be 0 (which is the starting point).

#Some pointers that might help with the code:

1. The function random.seed() may be used to make your results reproducible.
2. To represent clumsiness of falling down, generate a random number between 0 and 1. If that number is less than .001, set the steps value to Zero as is the rule.
3. You'll want to run the Dice game many times to calculate the probability of reaching step 60 in 100 rolls.




# Your Tasks

2. Estimate the probability of reaching step 60 in 100 rolls.

# Solution


```python
import matplotlib.pyplot as plt
import numpy as np
import sys

np.set_printoptions(linewidth=200)
np.set_printoptions(threshold=sys.maxsize)
```

```python


# Set seed for reproducibility
np.random.seed(123)

# Initialize all_walks to store the results of each simulation
all_walks = []

# Simulate the game 500 times
for i in range(500):
    random_walk = [0]
    for x in range(100):#represents the 100 rolls
        step = random_walk[-1]
        dice = np.random.randint(1, 7)#goes up to 7 because we want to go up to 6

        if dice <= 2:
            # Use max(0, step - 1) to ensure we don't go below ground level
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1, 7)

        # Implement clumsiness (0.1% chance)
        if np.random.rand() <= 0.001:
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Convert to a NumPy array to make data manipulation easier
np_aw = np.array(all_walks)

# Transpose the data so each column represents one walk
# This makes plotting much easier
np_aw_t = np.transpose(np_aw)

# Select the last row (the final step of each walk)
ends = np_aw_t[-1, :]

# Calculate probability
count_win = np.sum(ends >= 60)
prob = count_win / 500

print(f"Estimated Probability: {prob * 100}%")

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(np_aw_t)
plt.xlabel('Number of Rolls')
plt.ylabel('Step reached')
plt.title('Random Walks up the Empire State Building')
plt.show()
```

## Understanding the Logic

Let's break down the technical parts of this script:

np.random.randint(1, 7): This mimics a 6-sided die.

Note that in NumPy, the high value is exclusive, so (1, 7) gives us numbers 1 through 6.

The "Clumsiness" Check: We use np.random.rand(), which generates a decimal between 0 and 1. If it hits a value < 0.001, your progress resets to 0.

max(0, step - 1): This is a clean way to handle the rule that "the value of steps cannot be negative."

If 0 - 1 = -1, the max function picks 0 instead.

Transposing (.transpose()): Initially, our list "all_walks" is a list of lists where each row is a full walk. To plot 500 separate lines effectively, we flip the data so each "roll number" (0-100) is on the x-axis.

## Estimating the Probability


The "Monte Carlo" part of this happens at the very end. We look at the final step reached in all 500 simulations:

We count how many times the final step was 60 or higher.

We divide that by the total number of simulations (500).

In this specific run (with the seed 123), you’ll likely see a probability around 78.4%. This tells us that even with the risk of falling and the "downward" rolls, the "upward" rolls (3, 4, 5, and the 6 bonus) give you a strong chance of reaching the goal.

## Plotting a Histogram for Visual Clarity

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(123)

all_walks = []
for i in range(500):
    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1, 7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1, 7)

        if np.random.rand() <= 0.001:
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

np_aw_t = np.transpose(np.array(all_walks))
ends = np_aw_t[-1, :]

# Calculate probability of reaching step 60
prob_60 = np.sum(ends >= 60) / 500

# Create Line Plot
plt.clf()
plt.plot(np_aw_t, color='blue', alpha=0.1) # low alpha for better visualization
plt.title('All 500 Random Walks (Faded for clarity)')
plt.xlabel('Roll Number')
plt.ylabel('Step reached')
plt.savefig('random_walks.png')

# Create Histogram
plt.clf()
plt.hist(ends, bins=20, edgecolor='black', color='skyblue')
plt.axvline(60, color='red', linestyle='dashed', linewidth=2, label='Goal (Step 60)')
plt.title('Distribution of Final Steps reached')
plt.xlabel('Final Step Reached')
plt.ylabel('Number of Simulations (Frequency)')
plt.legend()
plt.savefig('histogram_ends.png')

print(f"Probability of reaching step 60: {prob_60}")
```

In the histogram, you are not plotting "probability" directly on the Y-axis. Instead, you are plotting Frequency vs. Final Step Reached.

## What if I plot the probability of reaching 60 steps with 100 rolls in a 3 D plot?  how would I do that?

```python
import plotly.graph_objects as go
import numpy as np

# Parameters for the 3D surface
roll_range = np.arange(50, 151, 10)   # X-axis
target_range = np.arange(40, 81, 5)   # Y-axis
simulations = 500

# Initialize a grid to store probabilities
z_data = np.zeros((len(target_range), len(roll_range)))

# Run simulations to fill the grid
for i, rolls in enumerate(roll_range):
    all_ends = []
    for _ in range(simulations):
        step = 0
        for _ in range(rolls):
            dice = np.random.randint(1, 7)
            if dice <= 2: step = max(0, step - 1)
            elif dice <= 5: step += 1
            else: step += np.random.randint(1, 7)
            if np.random.rand() <= 0.001: step = 0
        all_ends.append(step)

    # Calculate probability for each target step at this specific roll count
    for j, target in enumerate(target_range):
        prob = np.sum(np.array(all_ends) >= target) / simulations
        z_data[j, i] = prob

# Create the 3D Surface Plot
fig = go.Figure(data=[go.Surface(z=z_data, x=roll_range, y=target_range)])

fig.update_layout(
    title='Probability of Success: Rolls vs. Target Step',
    scene = dict(
        xaxis_title='Number of Rolls',
        yaxis_title='Target Step',
        zaxis_title='Probability'
    ),
    width=800, height=800
)

fig.show()
```

## How to Read This 3D Plot


When you run this code, you will see a Surface Map (often called a "Probability Landscape"):

The "High Ground" (Purple/Yellow): This is where the probability is close to 1.0 (100%). You'll see this where the Number of Rolls is high (150) and the Target Step is low (40).

The "Slope": This represents the "sweet spot" of the game. It shows how rapidly your chances drop as you either decrease your rolls or increase your target.

The "Valley": Where the probability drops toward 0. This happens if you try to reach a high step (80) with very few rolls (50).
