### Password Generator


Create a function that takes the length of the password and generates a random password of the same length with at least one character, one number and one symbol. Example: password(5) -> 'arg5!'

````python
import random
import string
import ipywidgets as widgets
from IPython.display import display, clear_output

````

````python
# 1. The password function
def generate_password(length):
    if length < 3:
        return "Length must be at least 3!"

    letters = string.ascii_letters
    digits = string.digits
    symbols = string.punctuation
    all_characters = letters + digits + symbols

    password_pool = [
        random.choice(letters),
        random.choice(digits),
        random.choice(symbols)
    ]

    remaining_length = length - 3
    password_pool += [random.choice(all_characters) for _ in range(remaining_length)]
    random.shuffle(password_pool)

    return "".join(password_pool)

````

````python
# 2. CREATE THE WIDGETS (This is the part Python couldn't find!)
slider = widgets.IntSlider(
    value=8,
    min=3,
    max=30,
    step=1,
    description='Length:',
    style={'description_width': 'initial'}
)

button = widgets.Button(
    description='Generate Password',
    button_style='success',
    icon='key'
)

output_area = widgets.Output()

````

````python
# 3. Define the click action
def on_button_clicked(b):
    with output_area:
        clear_output()
        password = generate_password(slider.value)
        print(f"🔑 Generated Password: {password}")

````

````python
# 4. Link the button click to the action
button.on_click(on_button_clicked)

````

````python
# 5. Display the dashboard
print("--- Interactive Password Generator ---")
display(slider, button, output_area)
````

### Email Slicer


- Create a function that takes as input an email address and returns a string that contains two items - the person's username and the domain name. Example: email_slicer('nevin@cnm.edu') -> ('nevin', 'cnm.edu')

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

# slicer
def email_slicer(email_address):
    """Splits an email address into a username and a domain name."""
    # Strip any accidental background spaces the user might type
    email = email_address.strip()

    # Validation check to ensure it is a real email format
    if "@" not in email or email.count("@") > 1:
        return "Invalid Email", "Please enter a valid email address (e.g., name@domain.com)"

    # Split the string at the '@' character
    username, domain = email.split("@")
    return username, domain

# Text input box for typing the email
email_input = widgets.Text(
    value='nevin@cnm.edu',
    placeholder='Type email address here',
    description='Email:',
    disabled=False
)

# Slider to control the text display size dynamically
size_slider = widgets.IntSlider(
    value=14,       # Default font size
    min=10,         # Minimum font size
    max=24,         # Maximum font size
    step=1,
    description='Font Size:',
    style={'description_width': 'initial'}
)

# Action button
slice_button = widgets.Button(
    description='Slice Email',
    button_style='info', # Gives it a nice blue color
    icon='cut'
)

# Area to display the results
output_display = widgets.Output()


# interactive button
def on_button_clicked(b):
    with output_display:
        clear_output() # Wipe old results

        # Call your core function using the text value from the widget
        user, domain = email_slicer(email_input.value)

        # Apply the slider's numeric value directly to the HTML font-size styling
        font_size = size_slider.value

        # Display the results using clean HTML formatting
        print("--- RESULTS ---")
        display(widgets.HTML(value=f"<p style='font-size:{font_size}px;'><b>Username:</b> {user}</p>"))
        display(widgets.HTML(value=f"<p style='font-size:{font_size}px;'><b>Domain:</b> {domain}</p>"))

# Link the button click to our action logic
slice_button.on_click(on_button_clicked)


#dashboard
print("--- Interactive Email Slicer Tool ---")
display(email_input, size_slider, slice_button, output_display)
````

### Acronym Creator


- Create a function that takes as input a string and returns an acronym that contains the first letter of each word in that string. Example: acronym_creator("My name is Nevin") -> "MNIN"

````python
def acronym_creator(text: str) -> str:
    """Takes a string and returns an uppercase acronym
    of the first letter of each word.
    """
    # Split the string by whitespace, take the first letter of each word,
    # convert to uppercase, and join them together.
    return "".join(word[0].upper() for word in text.split() if word)


# --- Example Usage ---
print(acronym_creator("Las achiras son hermosas"))  # Output: "MNIN"
print(acronym_creator("Nieva en Bariloche hoy"))  # Output: "ASAP"
````

### Dice Roller


- Create a function that takes as input a positive integer n. This function should simulate the rolling of a n-sided die 1000 times. It should create a seaborn countplot of the distribution of die rolls.

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


def simulate_die_rolls(n):
    """Simulates rolling an n-sided die 1000 times and plots the distribution."""
    if n < 1:
        raise ValueError("The number of sides 'n' must be a positive integer.")

    # 1. Simulate rolling the n-sided die 1000 times
    # np.random.randint is low-inclusive, high-exclusive, so we use (1, n + 1)
    rolls = np.random.randint(1, n + 1, size=1000)

    # 2. Set up the matplotlib figure size for better readability
    plt.figure(figsize=(10, 6))

    # 3. Create the seaborn countplot
    # We pass the rolls array directly as the data
    sns.countplot(x=rolls, palette="viridis")

    # 4. Add titles and labels to make the chart informative
    plt.title(
        f"Distribution of 1,000 Rolls of a {n}-Sided Die",
        fontsize=14,
        fontweight="bold",
    )
    plt.xlabel("Die Face", fontsize=12)
    plt.ylabel("Count", fontsize=12)

    # Clean up x-axis ticks so they show every integer face clearly
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    # 5. Display the plot
    plt.show()


# Example usage: Simulate a standard 6-sided die
simulate_die_rolls(40)

# Example usage: Simulate a 20-sided die (D20)
# simulate_die_rolls(20)
````

### Max of Three


- Create a function that takes as input three floats and returns the largest of the three. Do this without using the Python max() function! Example: max_of_three(5,10,3) -> 10

````python
def biggest_of_three(num1: float, num2: float, num3: float) -> float:
    """Returns the largest of three float numbers without using max()."""
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

print(biggest_of_three(5.0, 10.0, 3.0))  # Output: 10.0
````

### Age Calculator


- Create a function that takes as input someone's age in years and prints their age in days, hours and minutes. Example: age_calculator(33) -> "You are 12045 days old. You are 289080 hours old. You are 12592800 minutes old."

````python
# Version 1

def age_calculator(age_in_years: int) -> None:
    """Calculates and prints the age in days, hours, and minutes."""
    days = age_in_years * 365
    hours = days * 24
    minutes = hours * 60

    print(
        f"You are {days} days old. "
        f"You are {hours} hours old. "
        f"You are {minutes} minutes old."
    )
# edit the number in parenthesis below
age_calculator(67)
# Output: You are 12,045 days old, 289,080 hours old, 17,344,800 minutes old.
````
