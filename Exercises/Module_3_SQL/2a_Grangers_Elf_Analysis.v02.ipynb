# Python Exercises - Granger's Elf Analysis

Dr. Granger is interested in studying the relationship between the length of house-elves’ ears and aspects of their DNA. This research is part of a larger project attempting to understand why house-elves possess such powerful magic. She has obtained DNA samples and ear measurements from a small group of house-elves to conduct a preliminary analysis (prior to submitting a grant application to the Ministry of Magic) and she would like you to conduct the analysis for her. She has placed the file on the web for you to download, located ==> [here]( https://nyu-cds.github.io/courses/data/houseelf_earlength_dna_data.csv ) <==.

You might be able to do this analysis by hand in Excel, but counting all of those bases would be a lot of work, and besides, Dr. Granger seems to always get funded, which means that you’ll be doing this again soon with a much larger dataset. So, you decide to write a script so that it will be easy to do the analysis again.

Write a Python script that:

1. Imports the data into a data frame
2. Loops over the rows in the data frame
3. For each row in the dataset, checks to see if the ear length is large (>10 cm) or small (<=10 cm) and determines the GC-content of the DNA sequence (i.e., the percentage of bases that are either G or C)
4. Stores this information in a **new data frame** where the first column has the ID for the individual, the second column contains the string ‘large’ or the string ‘small’ depending on the size of the individuals ears, and the third column contains the GC content of the DNA sequence.
5. Prints the average GC-content for both large-eared elves and small-eared elves to the screen.
6. Exports the table of individual level GC values to a CSV (comma delimited text) file titled grangers_analysis.csv.

**BONUS**: Write functions that perform each of the steps above

# Solution - variant 1

Load in the modules


```python
import pandas as pd
import numpy as np
```

```python
url = 'https://nyu-cds.github.io/courses/data/houseelf_earlength_dna_data.csv'
url
```

### Request 1:

Imports the data into a data frame

```python
# Load the data natively from your Drive
df = pd.read_csv( url )
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Request 2 and 3:

Loops over the rows in the data frame - This was replaced with two function: one of the functions calculates the GC content, the other function converts the length of the ears in a string, large or small depending on the size.


For each row in the dataset, checks to see if the ear length is large (>10 cm) or small (<=10 cm) and determines the GC-content of the DNA sequence (i.e., the percentage of bases that are either G or C)

```python
# a vectorized approach
# this uses a regular expression in the .count() method
df['gc_content'] = df["dnaseq"].str.count(r'[GC]')/df["dnaseq"].str.len()*100
df
```

```python
# a vectorized approach
# uses numpy's .where() method
df['ear_size'] = np.where(df['earlength'] > 10, 'high', 'low')
df
```

### Request 4:

Stores this information in a new data frame where the first column has the ID for the individual, the second column contains the string ‘large’ or the string ‘small’ depending on the size of the individuals ears, and the third column contains the GC content of the DNA sequence.

```python
#creating a new df with only the information I need
grangers_df = df[['id', 'ear_size', 'gc_content']].copy()
grangers_df
```

### Request 5:

Prints the average GC-content for both large-eared elves and small-eared elves to the screen.

```python
# Group by the ear size and calculate the average GC content
# renames the series and converts to a data frame
grangers_df.groupby('ear_size')['gc_content'].mean().rename("mean GC content").reset_index()
```

### Request 6:


Exports the table of individual level GC values to a CSV (comma delimited text) file titled grangers_analysis.csv.

```python
# It saves the file directly to working directory in Colab
grangers_df.to_csv('grangers_analysis.csv', index=False)
print("\nFile successfully exported as 'grangers_analysis.csv'")
```
