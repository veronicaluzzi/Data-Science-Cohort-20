## Question: What are the names and composers on the tracks and invoice_items join table?

```python
# Select all common columns from tracks and invoice_items
%%script sqlite3 --column --header chinook.db
select *
from tracks
join invoice_items
on tracks.TrackId = invoice_items.TrackId
where GenreId = '1'
limit 10
```

### Question: What are the genre with ID's 1 and 2 from the join table between tracks and invoice_items?

#### AND


```python
# Select genre 1 and 2 from join table tracks + invoice_items
%%script sqlite3 --column --header chinook.db
select *
from tracks
join invoice_items
on tracks.TrackId = invoice_items.TrackId
where GenreId = '1' AND MediaTypeId = '2'
limit 10
```

### Question: similar as above but using 'or' instead of 'and'

#### OR


```python
# Select genre with ID's 1 and 2 from the join table between tracks and invoice_items
%%script sqlite3 --column --header chinook.db
select *
from tracks
join invoice_items
on tracks.TrackId = invoice_items.TrackId
where GenreId = '1' OR '2'
limit 10
```

### Question: What media name (song names) start with L?

#### LIKE (with % or _ wildcard)


```python
#to select songs with an L as the beginning of their name.
%%script sqlite3 --column --header chinook.db
select *
from tracks
where Name like 'L%'
```

### Question: What track ID's are between 30 and 40?

#### BETWEEN


```python
#to select Track Id between 30 and 40
%%script sqlite3 --column --header chinook.db

select *
from Tracks
where trackId BETWEEN 30 and 40
LIMIT 5
```

Question: what names of media that are between 480000 and 700000 msec?

#### LIMIT



```python
#to select milliseconds between 480000 and 700000
%%script sqlite3 --column --header chinook.db

select *
from tracks
where Milliseconds BETWEEN 480000 and 700000
LIMIT 10
```

### Sorting and Grouping


### Question: what is the name of media between 480000 and 700000, ordered by ascending order?

#### ORDER BY


```python
#to select milliseconds between 480000 and 700000, in ascending order
%%script sqlite3 --column --header chinook.db

select *
from tracks
where Milliseconds BETWEEN 480000 and 700000
order by Milliseconds
limit 10
```

### Question: What media milliseconds are distinct?

#### DISTINCT


```python
#to select distinct values in the milliseconds column
%%script sqlite3 --column --header chinook.db

select
distinct Milliseconds
from tracks
where Milliseconds BETWEEN 480000 and 700000
order by Milliseconds
limit 10

```

### Question: How many media names are there per composer listed in descending order?

#### GROUP BY



```python
%%script sqlite3 --column --header chinook.db
select
    Composer,
    COUNT(TrackId) AS SongCount
from tracks
group BY Composer
order BY SongCount DESC
limit 10;
```

### Aggregates


### Question: What is the length of the longest media in milisecond?

### Question: What is the length of the shortest media in milisecond?

### Question: What is the sum of all the media in miliseconds?

```python
%%script sqlite3 --column --header chinook.db
select sum(Milliseconds)
from tracks
;
```

### Question: What is the average of the media in milisecond?

#### AVG


```python
%%script sqlite3 --column --header chinook.db

select avg(Milliseconds)
from tracks
```

#### COUNT



```python
%%script sqlite3 --column --header chinook.db
select
    GenreId,
    COUNT(GenreId) AS GenCount
from tracks
group BY GenreId
order BY GenCount DESC
limit 10;
```

## Make some plots

Make some cool plots to go with your data. Write SQL queries to get ONLY the information you need for each plot. (Don't pull ALL the data and then just plot a few columns.)



```python
import sqlite3
import pandas as pd
import plotly.express as px

# 1. Connect and Query
conn = sqlite3.connect('chinook.db')
query = """
SELECT g.Name AS Genre, t.Milliseconds
FROM tracks t
JOIN genres g ON t.GenreId = g.GenreId
"""
df = pd.read_sql_query(query, conn)
conn.close()

```

```python
# 2. Create the Box Plot with Plotly Express
fig = px.box(df,
             x='Genre',
             y='Milliseconds',
             color='Genre',
             title='Distribution of Track Durations by Genre',
             labels={'Milliseconds': 'Duration (ms)', 'Genre': 'Music Genre'})

```

```python
# 3. Improve readability
fig.update_layout(xaxis_tickangle=-45)
fig.update_yaxes(type="log")
fig.show()
```
